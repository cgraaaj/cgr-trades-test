```python
df['Date_Time'] = pd.to_datetime(df['Date_Time'])
# Set the Date_Time column as the index
df.set_index('Date_Time', inplace=True)
ohlc_dict = {'LTP': ['first', 'max', 'min', 'last'],'OpenInterest': 'last'}
ohlc_df = df.resample('1T').agg(ohlc_dict)
ohlc_df.columns = ['Open', 'High', 'Low', 'Close', 'OpenInterest']

# Calculate the change in price (LTP) and OpenInterest
ohlc_df['OpenInterest_Change'] = ohlc_df['OpenInterest'].diff()
ohlc_df['LTP'] = ohlc_df['Close']
ohlc_df['Price_Change'] = ohlc_df['Close'].diff()

print(ohlc_df)
```
